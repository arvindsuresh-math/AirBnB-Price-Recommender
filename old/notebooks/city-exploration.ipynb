# City Exploration from Inside Airbnb Data

This notebook analyzes the scraped Inside Airbnb data to explore cities and their data availability.

```python
import pandas as pd
```

```python
df = pd.read_csv('../insideairbnb_all_data.csv')
print(f"Loaded {len(df)} rows of data.")
df.head()
```

```python
df['Country'].value_counts()
```

```python
df = df[df['Country'].isin(['United States', 'Canada', 'Australia', 'United Kingdom'])]
```

```python
df.info()
```

```python
country_dict = {country: df[df['Country'] == country] for country in ['United States', 'Canada', 'Australia', 'United Kingdom']}
```

```python
for country, country_df in country_dict.items():
    print(f"Country {country}: {len(country_df)} listings")
    print()
    print(country_df['City'].value_counts())
    print("="*20)
```
