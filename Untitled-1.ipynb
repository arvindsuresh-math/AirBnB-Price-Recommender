```python
from geopy.distance import geodesic
```

```python
public_coords = list(zip(stops.stop_lat, stops.stop_lon))  # (latitude, longitude)

def min_distance_to_trolley(row):
    point = (row['latitude'], row['longitude'])
    distances = [geodesic(point, trolley).km for trolley in public_coords]
    return min(distances)

DF['public_distance_km'] = DF.apply(min_distance_to_trolley, axis=1)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,5))

# Create a 1D density heatmap
sns.kdeplot(data=DF, x='public_distance_km', fill=True, cmap='Reds')

plt.title("Density of Distance to Nearest Public Transport")
plt.xlabel("Distance (km)")
plt.ylabel("Density")
plt.show()
```

```python
from sklearn.neighbors import BallTree

# Prepare coordinates (in radians for haversine distance)
airbnb_coords = np.radians(full_df[['latitude', 'longitude']].values)
stop_coords   = np.radians(stops[['stop_lat', 'stop_lon']].values)

# Build BallTree on stops
tree = BallTree(stop_coords, metric='haversine')

# Query nearest stop for each Airbnb
dist, ind = tree.query(airbnb_coords, k=1)

# Convert from radians to km (Earth radius  6371 km)
full_df['public_distance_km'] = dist.flatten() * 6371
```

```python
plt.figure(figsize=(8,5))

# Create a 1D density heatmap
sns.kdeplot(data=full_df, x='public_distance_km', fill=True, cmap='Reds')

plt.title("Density of Distance to Nearest Public Transport")
plt.xlabel("Distance (km)")
plt.ylabel("Density")
plt.show()
```

```python
import ast

# --- Step 1: Count number of properties per host ---
host_property_counts = (
    df.groupby('host_id')['listing_id']
      .nunique()  # number of unique listings per host
      .rename('num_properties')
)

# --- Step 2: Create has_cohost flag ---
def has_cohost_fn(x):
    if pd.isna(x) or str(x).strip() == '':
        return 0
    try:
        # if it's a list-like string such as "[123, 456]"
        cohosts = ast.literal_eval(x)
        return 1 if isinstance(cohosts, list) and len(cohosts) > 0 else 1
    except:
        return 1  # fallback if not parsable but non-empty

df['has_cohost'] = df['cohost_ids'].apply(has_cohost_fn)

# --- Step 3: Merge property counts back to main DataFrame ---
df = df.merge(host_property_counts, on='host_id', how='left')
```
